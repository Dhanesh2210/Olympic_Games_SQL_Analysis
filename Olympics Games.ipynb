#                               OLYMPIC GAMES DATA ANALYSIS

The Olympic Games are a global sporting event that bring together athletes from around the world to compete in a variety of sports. The rich history and data associated with the Olympics provide a valuable resource for analysis and insights into trends, performance, and participation over the years. In this notebook, we will explore the Olympic dataset to uncover interesting patterns and statistics.

## Importing the Dataset 

```python
import pandas as pd
olympics_df = pd.read_csv('dataset_olympics.csv')
```

```python
olympics_df
```

```python
olympics_df.info()
```

Filling the Null values in Height Weight columns with their median values

```python
olympics_df['Height'].fillna(olympics_df['Height'].median(), inplace=True)
olympics_df['Weight'].fillna(olympics_df['Weight'].median(), inplace=True)
```

```python
olympics_df.isnull()
```

Filling the null values in Medal column with 'No Medal'

```python
olympics_df['Medal'].fillna('No Medals', inplace=True)
olympics_df
```

Importing the Regions dataset

```python
noc_df = pd.read_csv('noc_region.csv')
noc_df
```

```python
noc_df.info()
```

Dropping the notes column as it is not required for analysis

```python
noc_df.drop(['notes'],axis=1,inplace=True)
noc_df
```

Renaming the columns to make them easy to understand and use in further steps

```python
noc_df.rename(columns= {'noc_region':'NOC','reg':'region'}, inplace=True)
noc_df
```

```python
olympics_df['Age'].info()
```

Filling the Null values in Age columns with their median values

```python
olympics_df['Age'].fillna(olympics_df['Age'].median(), inplace=True)
```

```python
olympics_df
```

Exporting the cleaned datasets into CSV files for analysis in SQL

```python
noc_df.to_csv('clean_noc.csv', index=False)
olympics_df.to_csv('clean_olympics.csv', index=False)
```
